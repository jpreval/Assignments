```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np

plt.figure(figsize=(30,30))

x = np.arange(-50,50,.1)

a = stats.norm.pdf(x,0,1)
b = stats.norm.pdf(x,0,10)
c = stats.norm.pdf(x,0,100)
d = stats.norm.pdf(x,0,1000)
e = stats.uniform.rvs(size=1000)
# print(len(x),len(a),
#      len(b),len(c),
#      len(d),len(e)
     
#      )
```

```python
# plt.figure(figsize=(200,200))
plt.plot(x,a,color='yellow', linestyle='solid', label = 'Normal: mean=0,Variance=1')
plt.plot(x,b,color='green', linestyle='solid', label = 'Normal: mean=0,Variance=10')
plt.plot(x,c,color='blue', linestyle='solid', label = 'Normal: mean=0,Variance=100')
plt.plot(x,d,color='red', linestyle='solid', label = 'Normal: mean=0,Variance=1000')
plt.plot(x,e,color='black', linestyle='solid', label = 'Uniform; a=0,b=1000')
plt.title('Mutiple graph')
plt.ylabel('ylabel')
plt.xlabel('xlabel')
plt.legend()

plt.show()
```
